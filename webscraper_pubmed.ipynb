# Webscraping and visualizing pubmed citations

My first attempt at scraping pubmed for information on crowding. To make it easier, we'll only look at articles in between 2010-now. 

## Request url and retrieve the info

```python
# Write function trasform keyword + date into url 
keyword = 'crowding'
from_ = '2010/01/01'
to = 'now'
```

```python
url = 'https://www.ncbi.nlm.nih.gov/pubmed?term=(crowding%5BTitle%2FAbstract%5D)%20AND%20(%222010%2F01%2F01%22%5BDate%20-%20Publication%5D%20%3A%20%223000%22%5BDate%20-%20Publication%5D)&dispmax=200'
```

```python
request = requests.get(url)
```

```python
pubmed_soup = BeautifulSoup(request.text, 'html.parser')
```

## Retrieve all papers on page

```python
papers_page = pubmed_soup.find_all('div', class_='rprt')
```

```python
first_paper = papers_page[0]
```

```python
title = first_paper.a.text
```

```python
pmid = int(first_paper.find('dl', class_='rprtid').text.split(' ')[1])
```

```python
authors = first_paper.find('p', class_='desc').text
first_author = authors.split(',')[0]
```

```python
journal = first_paper.find('span', class_='jrnl').text
```

```python
date = pd.to_datetime(re.split(r';|\.', first_paper.find('p', class_='details').text)[1])
```

```python
paper = {
    'pmid' : pmid, 
    'title' : title, 
    'date' : date,
    'first_author' : first_author, 
    'authors' : authors, 
    'journal' : journal
}
```

## XML parsing using BeautifulSoup

Now we'll use a frequently cited PMID to see whether we can extract all it's related citations. For this we will use the xml format of the html page. Via pubmed it is quite easy to access this information. 

```python
pmid = 22495300
citation_url = 'http://www.ncbi.nlm.nih.gov/pubmed/{}?report=xml'.format(pmid)
```

```python
request = requests.get(citation_url)
```

```python
# parse via BeautifulSoup xml
citation_soup1 = BeautifulSoup(request.text, 'xml')

# parse via BeautifulSoup html.parser
citation_soup2 = BeautifulSoup(request.text, 'html.parser')
```

```python
citation_soup = citation_soup1
```
